```python
# Web scraping da Câmara dos Deputados (Dados Abertos)
# Somente iDs
# Impressão: Arquivo .pdf das iDs

import pandas as pd
import requests

# Caminho do arquivo no computador
caminho_arquivo = r'C:\Users\Thairone Almeida\Google Drive\Thairone\Hashtag Treinamentos\Python Impressionador\Script\MBA_USP_ESALQ\Arquivo XML\iDs\link.xlsx'
dados = pd.read_excel(caminho_arquivo)

# Processar cada linha para baixar o PDF e salvá-lo com o ID correspondente como nome do arquivo
for indice, linha in dados.iterrows():
    if indice >= 30:  # Parar após 30 arquivos
        break

    url = linha['urlInteiroTeor']
    pdf_id = linha['id']

    # Baixar o PDF
    resposta = requests.get(url)
    
    if resposta.status_code == 200:
        # Salvar o arquivo PDF
        nome_arquivo_pdf = f"{pdf_id}.pdf"
        with open(nome_arquivo_pdf, 'wb') as arquivo:
            arquivo.write(resposta.content)
        print(f"Baixado e salvo: {nome_arquivo_pdf}")
    else:
        print(f"Falha ao baixar o PDF para o ID {pdf_id} da URL: {url}")
```
